```python
list_human= [sent_list[0] for sent_list in data_human.values]
list_reply= [sent_list[0] for sent_list in data_reply.values]
list_human
```

```python
human_token = []
reply_token = []

```

```python
for words in list_human:
    tokenized_words = word_tokenize(words)
    human_token.append(tokenized_words)
```

```python
len(human_token)
```

```python
for words in list_reply:
    tokenized_words = word_tokenize(words)
    reply_token.append(tokenized_words)
```

```python
len(reply_token)
```

```python
vector_represent_human=[]
for sent in human_token:
    vectors_human = [model[w] for w in sent if w in model.wv.vocab]
    vector_represent_human.append(vectors_human)
```

```python
len(vector_represent_human)
```

```python
vector_represent_reply=[]
for sent in reply_token:
    vectors_rep = [model[w] for w in sent if w in model.wv.vocab]
    vector_represent_reply.append(vectors_rep)
```

```python
len(vector_represent_reply)
```

```python
single_data = human_token + reply_token
```

```python
len(single_data)
```

```python
item = [eachitem[0] for eachitem in single_data]
    
```

```python
sentend=np.zeros((300,),dtype=np.float32) 
```

```python
for each_word in item:
    print(len(each_word))
   
```

```python
for tok_sent in vector_represent_reply:
    tok_sent[14:]=[]
    tok_sent.append(sentend)    
    
```

```python
# print(word_length)
for tok_sent in vector_represent_reply:
    if len(tok_sent)<15:
        for i in range(15-len(tok_sent)):
            tok_sent.append(sentend) 
```
